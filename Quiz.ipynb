```
# import modules 
import spacy
import wikipedia
import spacy
import pprint
import random
from nltk import sent_tokenize
from random import randint

# load spacy large library
nlp = spacy.load('en_core_web_lg')
```

## Select one of the topics from the list above and paste it the cell below as topic

```
topic = "Cricket"
summary = wikipedia.summary(topic)
```

```

def replace_str_index(text, start_index, end_index, replacement='____'):
    return '%s%s%s'%(text[:start_index],replacement,text[end_index:])


def get_unique_options(options_list):
  options_list = [op[0] for op in options_list]
  answer = options_list[0]
  options_list = list(dict.fromkeys(options_list))
  list_len = len(options_list)
  if(list_len == 1):
    return []
  if(list_len > 3):
    options_index  = random.sample(range(1, list_len), 3)
    options = [options_list[i] for i in options_index]
  else:
    options_index  = random.sample(range(1, list_len), list_len-1)
    options = [options_list[i] for i in options_index]
  
    
  options.append(answer)
  random.shuffle(options) 
  return options


def detect(text):
  doc = nlp(text)
  entities = {}
  for e in doc.ents:
    if(e.label_ in entities):
      entities[e.label_].append((e.text, e.start_char, e.end_char))
    else:
      entities[e.label_] = [(e.text, e.start_char, e.end_char)]
  
  new_str = ""
  # selected_entity = "GPE"
  selected_entity = random.choice(list(entities.keys()))

  if(selected_entity in entities):
    random.shuffle(entities[selected_entity]) 
    index_of_selected_answer = 0
    
    answer = entities[selected_entity][index_of_selected_answer][0]
    new_str = replace_str_index(text, entities[selected_entity][index_of_selected_answer][1], entities[selected_entity][index_of_selected_answer][2])
    sentences = sent_tokenize(new_str)
    for sentence in sentences:
      if("____" in sentence):
        question = sentence
        break


    options = get_unique_options(entities[selected_entity])
  return question, answer, options
ques, answer, options = detect(summary)
print("Question:\n")
pprint.pprint(ques)
print("\n")
if(len(options)>0):
  print("Options:\n ")
  print(options)

```
